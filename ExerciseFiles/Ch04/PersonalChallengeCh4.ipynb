## 2. Which country has won the most men's gold medals in singles badminton over the years (Sort by: Player's Names)

```python
bm = oo[(oo.Sport == 'Badminton') & (oo.Event == 'singles') & (oo.Medal == 'Gold') & (oo.Gender == 'Men')].sort_values('Athlete')
bm.head()
```

```python
bm.NOC.value_counts()
```

## 4. Display the male gold medal winners for the 100m track and field sprint event over the years (Sort by: Starting with most recent)(Output: City, Edition, Athlete, and Country Represented)

```python
mgm = oo[(oo.Gender == 'Men') & (oo.Event == '100m') & (oo.Medal == 'Gold') & (oo.Sport == 'Athletics')]
mgm = mgm[['City', 'Edition', 'Athlete', 'NOC']].sort_values(by='Edition',ascending=False)
mgm
```
