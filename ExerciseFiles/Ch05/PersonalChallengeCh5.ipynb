# The Main Challenge

## Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using both Matplotlib and Seaborn

```python
cme = oo[(oo.NOC == 'CHN') & (oo.Edition == 2008)]
cme.head()
```

```python
cme.Gender.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='Gender',data=cme)
```

## Plot the number of gold, silver, and bronze medals for each gender.

```python
sns.countplot(x='Medal',data=cme, hue='Gender')
```

## How can you give the data more meaning? Is there anything else you can change to make it more intuitive?

```python
cme.Gender.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='Gender',data=cme, palette='bwr')
```

```python
sns.countplot(x='Medal',data=cme, hue='Gender', palette='bwr',order=['Gold', 'Silver', 'Bronze'])
```
